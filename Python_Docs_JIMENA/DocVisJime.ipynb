```python
def tiempo(t):
    rel4= t.period.str.split(pat=' ',expand=True)
    rel4[['num','word']]=t.period.str.split(pat=' ', expand=True)
    rel4.loc[rel4['word'] == 'month', 'word'] = '30'
    rel4.loc[rel4['word'] == 'day', 'word'] = '1'
    rel4.loc[rel4['word'] == 'week', 'word'] = '7'
    rel4 = (rel4.astype({
        'num':  'int8',
        'word': 'int8',}))
    rel4['Period'] = rel4['num'] * rel4['word']
    return rel4
```

```python
for i in range(ind):#El rango funciona automatico desde cero hasta entermos menores al colocado
    tab = dff[dff['name'] == rel3[i]] #rel3 tiene índices comenzando del 0
    name=rel3[i]#Guardamos el nombre en un variable para así llamar su archivo
    LT=tiempo(tab)
    TAB=pd.concat([LT.Period, tab], axis=1)
    TAB.pop('OB')
    TAB.pop('name')
    s=TAB.style.bar(subset=['Period'], color='#20C9D0').background_gradient(gmap=TAB['painful'], vmin=1, vmax=10, cmap=cm, axis=0, subset=['bodySite'])
    s.hide(["painful","period"], axis=1)
    dfi.export(s, name)
    print (s)
```
